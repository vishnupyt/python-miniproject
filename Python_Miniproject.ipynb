```python
# Task: Inventory Management System
# 1. Store items in a list with their quantities. Add functions to:
# 2. Add new items
# 3. Update quantities
# 4. Remove items when quantities reach zero
```

```python
inventory = []

def add_new_items(item_name, quantity):
    item = {"name": item_name, "quantity": quantity}
    inventory.append(item)
    print(f"{item_name} has been added to the inventory with quantity {quantity}")

def update_quantities(item_name, additional_quantity):
    for item in inventory:
        if item["name"] == item_name:
            item["quantity"] += additional_quantity
            print(f"The quantity of {item_name} has been updated to {item['quantity']}")
            return
    print(f"{item_name} is not in the inventory, so cannot update quantity.")

def remove_items(item_name):
    for item in inventory:
        if item["name"] == item_name:
            inventory.remove(item)
            print(f"{item_name} has been removed from the inventory")
            return
    print(f"{item_name} is not in the inventory")

def buy_item(item_name, quantity):
    for item in inventory:
        if item["name"] == item_name:
            if item["quantity"] >= quantity:
                item["quantity"] -= quantity
                print(f"You bought {quantity} of {item_name}. Remaining quantity: {item['quantity']}")
                if item["quantity"] == 0:
                    inventory.remove(item)
                    print(f"{item_name} is now out of stock and removed from inventory.")
                return
            else:
                print(f"Not enough {item_name} in stock. Available quantity: {item['quantity']}")
                return
    print(f"The item '{item_name}' is over and not available in inventory.")

print("WELCOME TO MY INVENTORY MANAGEMENT SYSTEM")

while True:
    print("\nMenu:")
    print("1. Add new items")
    print("2. Update quantities")
    print("3. Remove items when quantities reach zero")
    print("4. Buy an item")
    print("5. Exit")

    function = input("Enter function (1,2,3,4,5): ")

    if function == "1":
        number_of_items = int(input("Enter number of items: "))
        for _ in range(number_of_items):
            item_name = input("Enter item name: ")
            quantity = int(input(f"Enter quantity for {item_name}: "))
            add_new_items(item_name, quantity)
        print("Current Inventory:", inventory)

    elif function == "2":
        item_name = input("Enter item name: ")
        additional_quantity = int(input("Enter additional quantity to add: "))
        update_quantities(item_name, additional_quantity)
        print("Current Inventory:", inventory)

    elif function == "3":
        item_name = input("Enter item name: ")
        remove_items(item_name)
        print("Current Inventory:", inventory)

    elif function == "4":
      item_name = input("Enter item name to buy: ")
      quantity = int(input("Enter quantity to buy: "))
      buy_item(item_name, quantity)
      print("Current Inventory:", inventory)

    elif function == "5":
        print("Exiting the Inventory System, Good Bye!")F
        break

    else:
        print("Invalid option. Please choose a valid function (1, 2, 3, 4, or 5).")
```
